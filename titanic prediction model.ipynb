```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data=pd.read_csv("../input/titanic/train.csv")
```

```python
df_x=data.drop(['Survived'],axis=1)
```

```python
df_x.head(2)
```

```python
df_y=data['Survived']
df_y.head(2)
```

```python
df_x.iloc[:,9:10]=df_x.iloc[:,9:10].apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
df_x.iloc[:,10:11]=df_x.iloc[:,10:11].apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
from sklearn.impute import SimpleImputer
imput=SimpleImputer(missing_values=np.nan,strategy='mean',verbose=0)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_x['Name']=le.fit_transform(df_x.Name)
df_x['Sex']=le.fit_transform(df_x.Sex)
df_x['Ticket']=le.fit_transform(df_x.Ticket)
df_x['Cabin']=le.fit_transform(df_x.Cabin)
df_x['Embarked']=le.fit_transform(df_x.Embarked)
```

```python
df_x.head(2)
```

```python
x=df_x.iloc[:,:].values
imput=imput.fit(x[:,:])
x=imput.transform(x[:,:])
x=pd.DataFrame(x)
```

```python
x=x.rename(columns={0:'PassengerId',1:'Pclass',2:'Name',3:'Sex',4:'Age',5:'SibSp',6:'Parch',7:'Ticket',8:'Fare',9:'Cabin',10:'Embarked'})
```

```python
from sklearn.ensemble import RandomForestClassifier
re=RandomForestClassifier(n_estimators=100)
```

```python
x3=x.drop(['Cabin','Name'],axis=1)
```

```python
x3=x3.drop(['Ticket'],axis=1)
x3.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x3,df_y,test_size=0.2,random_state=0)
```

```python
re.fit(x_train,y_train)
```

```python
re.predict(x_test)
```

```python
re.score(x_test,y_test)
```

```python
ress=pd.read_csv('../input/titanic/test.csv')
```

```python
ress=ress.drop(['Cabin'],axis=1)
ress['Sex']=le.fit_transform(ress.Sex)
ress['Embarked']=le.fit_transform(ress.Embarked)
ress['Ticket']=le.fit_transform(ress.Ticket)
ress['Name']=le.fit_transform(ress.Name)
ress.head()
```

```python
yy=ress.iloc[:,:].values
imput=imput.fit(yy[:,:])
yy=imput.transform(yy[:,:])
yy=pd.DataFrame(yy)
```

```python
yy=yy.rename(columns={0:'PassengerId',1:'Pclass',2:'Name',3:'Sex',4:'Age',5:'SibSp',6:'Parch',7:'Ticket',8:'Fare',9:'Embarked'})
yy.head()
```

```python
yy=yy.drop(['Name','Ticket'],axis=1)
```

```python
yy.head()
```

```python
sub=pd.DataFrame(re.predict(yy))
```

```python
sm=yy.drop(['SibSp','Pclass','Sex','Age','Parch','Fare','Embarked'],axis=1)
sm.head(10)
```

```python
sub=sub.rename(columns={0:'Survived'})
```

```python
sb=pd.concat([sm,sub],axis=1)
```

```python
sb['PassengerId'] = sb['PassengerId'].apply(lambda x: np.int64(x))
sb.head()
```
